```
#Filter out product codes for train Data
train_productcode =train[['P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]

#Filter out product codes for test Data
test_productcode =test[['P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]
```

```
#check shape of train and test dataset
print(train.shape, test.shape)
```

#Convert product Code columns into one columns 

```
##basically it customers*productcode
train = train.melt(id_vars=train.columns[:8], value_vars=train_productcode, var_name = "PCODE", value_name="Label" )

#basically it customers*productcode
test = test.melt(id_vars=test.columns[:8], value_vars=test_productcode, var_name = "PCODE", value_name="Label" )
```

```
#After Converting product Code columns into one columns the shape has changed
print(train.shape, test.shape)
```

```
#Create X variable That would be used to ID and PCODE in train data
train['join']='x'

#Create X variable That would be used to ID and PCODE in test data
test['join']='x'
```
